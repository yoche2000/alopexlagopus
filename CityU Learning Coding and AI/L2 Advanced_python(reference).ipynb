# Demonstration - Finding the Value of Pi using Monte Carlo with Numpy

```
import numpy as np
```

Generate Monte Carlo samples

```
m = int(1e6) #Monte Carlo sample size

x = np.random.uniform(0,1,m)
y = np.random.uniform(0,1,m)
```

Getting Monte Carlo estimate of $\pi$

```
p =  x**2+y**2 <= 1.0**2

pi = np.mean(p)*4

print('The Monte Carlo estimate of pi is: %.2f' % pi)
```

Error of the estimate within 95% Confident Interval

```
sigma_sq = (1/(m-1))*np.sum( (4*p - pi)**2 )
err = np.sqrt(sigma_sq**2/m)*1.96
print('Error within 95%% CI is: %.3f' % err )
```

# **Note about numpy array**

Changing an element of b_list won't affect the elements of a_list

```
a_list = list(range(6))
print(a_list)
b_list = a_list[1:4]
print(b_list)
```

```
b_list[0] = 1000
```

```
print(a_list)
print(b_list)
```

But changing an element of b_array will also change the corresponding element of a_array

```
a_array = np.arange(6)
print(a_array)
b_array = a_array [1:4]
print(b_array)

```

```
b_array [0] = 1000
```

```
print(a_array)
print(b_array)
```

You need to create a new instant of an array when creating b_array if you don't want to alter a_array when altering b_array

```
a_array = np.arange(6)
print(a_array)
b_array  = a_array [1:4].copy() # or b_array  = np.array(a_array [1:4])

print(b_array)

```

```
b_array [0] = 1000
```

```
print(a_array)
print(b_array)
```

# **Rolling window**

```
df = pd.DataFrame(historical_data)
del df['date']
df.rename(columns={'formatted_date':'date'}, inplace=True)
df.index = pd.DatetimeIndex(df['date'])
df.tail()
```

```
df['close_7days_sma'] = df['close'].rolling(7).mean()
df['close_20days_sma'] = df['close'].rolling(20).mean()
```

```
df['20std'] = df['close'].rolling(20).std()

df['bollinger_upper'] = df['close_20days_sma'] + 2*df['20std']
df['bollinger_lower'] = df['close_20days_sma'] - 2*df['20std']
```

```
def wma(x):
    n = x.shape[0] #no. of days
    w = np.arange(1,n+1).astype('float64')
    w /= np.sum(w) #normalised weights 
    #Write your code here (1~2 lines)

    return np.sum(w*x)

df['wma20'] = df['close'].rolling(20).apply(wma,raw=True)
df['wma20'].tail()
```

```
df.loc['2018-09-01':'2019-01-01' ,['close', 'close_20days_sma', 'wma20', 'bollinger_upper','bollinger_lower']].plot(
    figsize=(12,7),
    kind='line',
    subplots=False,
    title = 'Bollinger Bands',
    legend=True
)
```
