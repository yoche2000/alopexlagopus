# Run the below code sections to import Tencent stock data from Yahoo Finance

1. This will install the yahoofinancials library for importing the stock data, reference: https://github.com/JECSand/yahoofinancials

```
!pip install yahoofinancials
```

2. Import the Tencent (0700.HK) stock price, you can change the end_date to the date of today if you like. Note the data type of historical_data.

```
from yahoofinancials import YahooFinancials as yf

ticker = '0700.HK'
start_date = '2009-01-01'
end_date = '2019-01-01'
freq = 'daily'

yahoo_financials = yf(ticker)
historical_data = yahoo_financials.get_historical_price_data(start_date, end_date, freq)[ticker]['prices']


print(historical_data)

```
